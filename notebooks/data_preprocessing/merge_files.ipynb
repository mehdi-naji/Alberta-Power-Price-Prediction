```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
supply_load_price = pd.read_csv('../../data/processed/supply_load_price.csv', parse_dates=['date'], index_col='date')
supply_load_price = supply_load_price.asfreq('H')
supply_load_price = supply_load_price.sort_values(by='date')
supply_load_price
```

```python
intertie_df = pd.read_csv('../../data/processed/intertie.csv', parse_dates=['date'], index_col='date')
intertie_df = intertie_df.asfreq('H')
intertie_df = intertie_df.sort_values(by='date')
intertie_df.head(2)
```

```python

intertie_df[intertie_df['atc_sk_export'].isna()].index
```

```python
merged_df = pd.merge(supply_load_price, intertie_df, left_index=True, right_index=True)
final_merge = merged_df


final_merge[final_merge['price'].isna()].index
```

```python
final_merge.to_csv('../../data/processed/final_df.csv')
final_merge.tail(5)
```

```python
import pandas as pd

# Assuming your DataFrame is named 'df'

# Create a reference DataFrame with a complete set of hourly timestamps
min_date = final_merge.index.min()
max_date = final_merge.index.max()
reference_index = pd.date_range(min_date, max_date, freq='H')
reference_df = pd.DataFrame(index=reference_index)

# Merge the original DataFrame with the reference DataFrame
merged_df = pd.merge(reference_df, final_merge, how='left', left_index=True, right_index=True)

# Check for missing values (NaN) in the merged DataFrame, which indicates missing hours
missing_hours = merged_df[merged_df.isnull().any(axis=1)].index

if missing_hours.empty:
    print("No missing hours in the data.")
else:
    print("Missing hours:")
    print(missing_hours)
```
